```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By


import time
from bs4 import BeautifulSoup

import pandas as pd
```

```python
driver = webdriver.Chrome()
driver.get('https://www.youtube.com/@jacksfilms/videos')

for _ in range(60):
    driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
    time.sleep(3)
```

```python
html = driver.page_source
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
videos = soup.find_all('div', {'id':'dismissible'})
```

```python
video_list = []

for video in videos:
    data_dict = {}
    data_dict['title'] = video.find('a', {'id':'video-title-link'}).text
    
    meta = video.find('div', {'id':'metadata-line'})
    data_dict['views'] = meta.find('span', {'class':'inline-metadata-item style-scope ytd-video-meta-block'}).text
    data_dict['date'] = meta.find_all('span', {'class':'inline-metadata-item style-scope ytd-video-meta-block'})[1].text
    
    data_dict['url'] = 'https://www.youtube.com' + video.find('a',{'id':'thumbnail'})['href']
    
    video_list.append(data_dict)
```

```python
len(video_list)
```

```python
df = pd.DataFrame(video_list)
```

```python
df
```

```python
def view_count(df):
    if 'K' in df:
        df = float(df.split('K')[0])*1000
    elif 'M' in df:
        df = float(df.split('M')[0])*1000000
    return df
```

```python
df['cleaned_views'] = df['views'].apply(view_count)
```

```python
df['cleaned_views'] = df['cleaned_views'].astype('int')
```
